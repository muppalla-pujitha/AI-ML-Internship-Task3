```python
# ================================
# Task 3: Exploratory Data Analysis (EDA)
# Dataset: iris.data.csv
# ================================

# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("iris.data.csv")

# Add Column Names
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Display First Rows
print("First 5 Rows of Dataset:")
print(df.head())

# Dataset Information
print("\nDataset Info:")
print(df.info())

# Statistical Summary
print("\nDataset Description:")
print(df.describe())

# ================================
# 1. Histogram - Distribution of Numerical Features
# ================================
print("\nPlotting Histograms...")
df.hist(figsize=(10,8))
plt.suptitle("Histogram of Iris Features")
plt.show()

# ================================
# 2. Count Plot - Categorical Feature
# ================================
print("\nPlotting Count Plot...")
sns.countplot(x='species', data=df)
plt.title("Species Count")
plt.show()

# ================================
# 3. Box Plot - Outlier Detection
# ================================
print("\nPlotting Box Plot...")
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['sepal_length','sepal_width','petal_length','petal_width']])
plt.title("Box Plot for Outlier Detection")
plt.show()

# ================================
# ================================
# Correlation Heatmap (FIXED)
# ================================
print("\nPlotting Correlation Heatmap...")

# Select only numeric columns
numeric_df = df[['sepal_length','sepal_width','petal_length','petal_width']]

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ================================
# 5. Pair Plot (Optional but Good)
# ================================
print("\nPlotting Pair Plot...")
sns.pairplot(df, hue='species')
plt.show()

# ================================
# INSIGHTS
# ================================
print("\nINSIGHTS:")
print("1. Petal length and petal width are highly correlated.")
print("2. Setosa species is clearly separable from others.")
print("3. Some outliers are present in sepal width.")
print("4. Petal features are most important for prediction.")

# ================================
# SUMMARY
# ================================
print("\nSUMMARY:")
print("Dataset has 150 rows and 5 columns.")
print("No missing values found.")
print("Petal length and petal width are key features for classification.")
print("EDA helps understand data patterns and relationships.")
```
